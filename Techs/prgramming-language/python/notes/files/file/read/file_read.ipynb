## 判断是否为gif格式文件

```python
# -*- coding: utf-8 -*-

def is_gif(fname):
    f = open(fname, 'br')
    first4byte = tuple(f.read(4))
    return first4byte == (0x47, 0x49, 0x46, 0x38)

# 如上函数是通过读取一个文件前面4个字节来判断是否为gif格式的文件。
```

## 找到文件最后一行

```python
with open("textfile.txt", 'r') as f:
    lines = f.readlines()
    first = lines[0].strip()
    last = lines[-1].strip()

    print("firt line = {} and second line = {}".format(first, last))

for line in lines:
    print(line.strip())
```

```python
def readlastline(f):
    f.seek(-2, 2)              # Jump to the second last byte.
    while f.read(1) != b"\n":  # Until EOL is found ...
        f.seek(-2, 1)          # ... jump back, over the read byte plus one more.
    return f.read()            # Read all data from this point on.
    # 看起来不好懂是不？用中文翻译一下就是，首先跳到文件末尾，然后逐个往前找，直到找到倒
    # 数第二行的EOL，从那个位置读取的就是最后一行
with open("textfile.txt", "rb") as f:
    first = f.readline()
    last = readlastline(f)
    print("firt line = {} and second line = {}".format(first, last))
```
