## 离散均匀随机分布(discrete uniform distribution)

```python
import statistics
import numpy as np
import matplotlib.pyplot as plt

arr = np.random.randint(0, 100, 10000)
print(statistics.mean(arr))
print(statistics.stdev(arr))

fig, axes = plt.subplots(2, 1)
axes[0].hist(arr, bins=100)
axes[1].plot(arr)
```

## 连续均匀分布(continuous uniform distribution)

```python
import numpy as np
import matplotlib.pyplot as plt

# 符合[0, 1)的连续均匀分布的随机数
#arr = np.random.random(10000)

# 创建mean-std, mean+std的连续均匀分布数据
lower_bound = 10 - 1
upper_bound = 10 + 1
arr = np.random.uniform(lower_bound, upper_bound, 10000)
print(statistics.mean(arr))
print(statistics.stdev(arr))

fig, axes = plt.subplots(2, 1)
axes[0].hist(arr, bins=100, density=False)
axes[1].plot(arr)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#符合标准正态分布的随机数
#arr = np.random.randn(10000)

#符合特定正态分布的随机数
arr = np.random.normal(loc=10, scale=1.0, size=10000)
print(statistics.mean(arr))
print(statistics.stdev(arr))

fig, axes = plt.subplots(2, 1)
axes[0].hist(arr, bins=100)
axes[1].plot(arr)
```
