## 数据透视:`pivot()`

测试`pivot_table()`：

```python
import numpy as np
import pandas as pd

date = pd.date_range('20220101', periods=8)

types = ['A', 'B', 'C', 'D']
df_pivot_table = pd.DataFrame({
    'date' : date,
    'class1' : [types[x] for x in np.random.randint(0, 4, size=8)],
    'class2' : [types[x] for x in np.random.randint(0, 4, size=8)],
    'score' : np.random.random(8)
})

df_pivot_table
```

```python
# 使用`pivot_table`的时候其实需要指定对应的聚合函数，默认为`mean`
df_pivot_table.pivot_table(index='class1', columns='class2', values='score')
```

测试`pivot()`：

```python
import numpy as np
import pandas as pd

date = pd.date_range('20220101', periods=6)

types = ['A', 'B', 'C', 'D']
df = pd.DataFrame({
    'date' : date,
    'class' : [types[x] for x in np.random.randint(0, 4, size=6)],
    'score' : np.random.random(6)
})

df
```

```python
# 数据透视，将`date`列转换为行，`variables`列转换为列，`score`列转换为值，查看三列之间的关系
pivot_df = df.pivot(index='date', columns='class', values='score')
pivot_df
```

```python
# 数据透视之后通过stack，看起来复原了，但实际上是一个`Series`对象
pivot_stack_df = pivot_df.stack()
pivot_stack_df
```

```python
# 逆操作就是`melt()`函数，先需要把`index`转换为普通列
reset_index_df = pivot_df.reset_index()

df_melt = reset_index_df.melt(id_vars=['date'], value_vars=['A', 'C', 'D'])
df_melt
```

## melt()

```python
import numpy as np
import pandas as pd

date = pd.date_range('20220101', periods=6)

types = ['A', 'B', 'C', 'D']
df = pd.DataFrame({
    'date' : date,
    'class' : [types[x] for x in np.random.randint(0, 4, size=6)],
    'score' : np.random.random(6)
})

df
```

```python
df_melt = df.melt(id_vars=['date'], value_vars=['class', 'score'])
df_melt
```

```python
df_melt.pivot(index='date', columns='variable', values='value')
```

## 堆叠：stack()/unstack()

```python
import numpy as np
import pandas as pd

date = pd.date_range('20220101', periods=6)

types = ['A', 'B', 'C', 'D']
df = pd.DataFrame({
    'date' : date,
    'classs' : [types[x] for x in np.random.randint(0, 4, size=6)],
    'score' : np.random.random(6)
})

df
```

```python
# 之前的DataFrame，会堆叠为Series
df_stack = df.stack()
df_stack
```

```python
# df_stack 为单索引，压缩为`Series`类型
df_stack.info()
```

```python
df_stack.unstack()
```
