```python
import os
import pandas as pd
import matplotlib.pyplot as plt

cols = ['rnti', 
        'ulPreSchedData.highestClassPriority',
        'ulFdSchedData.maxHarqTxNum', 'ulFdSchedData.txNumber', 
        'ulHarqRxData.crcResult', 'ulHarqRxData.dtx',
        'ulLaAverageSinr.rrmInstSinr']

path = r'C:\N-20L6PF1F2MV8-Data\lianbche\Documents\2024\04\17-03773022\NPH10523E'
df = pd.read_csv(
        os.path.join(
        path,
        'UTF_20240405T212012_29.28.194.218_l2Tracer_NPH10523E_29.28.194.218_20240405T212012.202339_L2_COMPONENT_0000_1_ul.csv'),
        usecols=cols)

fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)
df.groupby('rnti')['ulHarqRxData.crcResult'].value_counts().unstack().plot(ax=axes[0], kind='bar', stacked=True, figsize=(8, 4), title='crcResult')
df.groupby('rnti')['ulHarqRxData.dtx'].value_counts().unstack().plot(ax=axes[1], kind='bar', stacked=True, figsize=(8, 4), title='dtx')

fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)
tx5df = df[df['ulFdSchedData.txNumber'] == 5]
tx5df.groupby('rnti')['ulHarqRxData.crcResult'].value_counts().unstack().plot(ax=axes[0], kind='bar', stacked=True, figsize=(8, 4), title='crcResult for 5th TX')
tx5df.groupby('rnti')['ulHarqRxData.dtx'].value_counts().unstack().plot(ax=axes[1], kind='bar', stacked=True, figsize=(8, 4), title='dtx for 5th TX')

tx5df.groupby('rnti')['ulPreSchedData.highestClassPriority'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(8, 4), title='highestClassPriority for 5th TX')

#print(df['ulFdSchedData.maxHarqTxNum'].dropna().unique())
#[5.]
#crnti = 40952
#df[df['rnti'] == crnti]['sinr_[0]'].plot(figsize = (6, 2), title='CRNTI: {}'.format(crnti))
```
