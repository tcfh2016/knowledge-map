step 1: import 必要的库

```python
import pandas as pd
```

step 2: 使用read_excel阅读excel数据文件，默认读取第一个sheet

```python
pd.read_excel("fund_data.xlsx")
```

step 3: 使用sheet_name来指定要读取的sheet

```python
pd.read_excel("fund_data.xlsx", sheet_name=1)
```

```python
pd.read_excel("fund_data.xlsx", sheet_name="股票基金")
```

step 4: 使用sheet_name来指定多个sheet


```python
pd.read_excel("fund_data.xlsx", sheet_name=[0, "股票基金"])
```

step 5: 获取所有sheet

```python
df = pd.read_excel("fund_data.xlsx", sheet_name=None)
print(df['混合基金'])
```
