## 多副图

```python
import numpy as np
import matplotlib.pyplot as plt
# matplotlib.rcParams['axes.unicode_minus'] = False


t = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)

plt.figure()
plt.plot(x)

plt.figure()
plt.plot(y);
```

## 多副子图

```python
plt.subplot(1, 2, 1)
plt.plot(x)
plt.subplot(1, 2, 2)
plt.plot(y);
```

## 多条线

```python
plt.figure()
plt.plot(x)
plt.plot(y)
```
