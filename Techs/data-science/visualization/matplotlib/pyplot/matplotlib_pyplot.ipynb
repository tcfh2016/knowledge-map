# 常规设置

## 设置画布

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5), facecolor='yellow', edgecolor='r')
ax = fig.subplots()

x = np.linspace(0, 10, 10)
y = np.random.randint(0, 100, 10)

ax.plot(x, y, color='r', marker='v', mfc='b')
```

## 添加文本标签

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = np.random.randint(0, 100, 10)

plt.plot(x, y, color='r', marker='v', mfc='b')

for a, b in zip(x, y):
    plt.text(a, b+3, b, ha='center', va='bottom', fontsize=9)
```

## 设置图例

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

fig = plt.figure(figsize=(10, 5), edgecolor='r')

df=pd.read_excel('体温.xls')

x=df['日期']
y=df['体温']
plt.title('体温数据')
plt.xlabel('2020年2月')
plt.ylabel('度数')
plt.legend(('基础体温',), bbox_to_anchor=(1.05, 1), borderaxespad=0, fontsize=12)

plt.plot(x, y, color='b', marker='o', mfc='r')

```

## 调整图表和画布间距

```python
plt.subplots_adjust(left=1, bottom=1, right=2, top=2, wspace=0, hspace=0)

plt.plot(x, y, color='b', marker='o', mfc='r')
```

# 设置坐标轴

## 坐标轴标题

- `ax.set_xlabel('xxx')`/`plt.xlabel('xxx')`设置x轴标题
- `ax.set_ylabel('xxx')`/`plt.ylabel('xxx')`设置y轴标题
- `mpl.rcParams['font.sans-serif'] = ['SimHei']`/`plt.rcParams['font.sans-serif'] = ['SimHei']`解决中文乱码问题
- `mpl.rcParams['axes.unicode_minus'] = False``/`plt.rcParams['axes.unicode_minus'] = False`解决负号不显示问题

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

# `fig`虽然没有直接使用到，但是它是一个`Figure`对象，可以用来设置图形的一些属性
fig = plt.figure(figsize=(10, 5), facecolor='yellow', edgecolor='r')

# 此时画图没有使用到`fig`
x = np.linspace(0, 10, 10)
y = np.random.randint(0, 100, 10)
plt.plot(x, y, color='b', marker='o', mfc='r')

plt.xlabel('x轴坐标')
plt.ylabel('一些随机值')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.sans-serif'] = ['SimHei']

fig = plt.figure(figsize=(10, 5), facecolor='yellow', edgecolor='r')
ax = fig.subplots()

x = np.linspace(0, 10, 10)
y = np.random.randint(0, 100, 10)
ax.plot(x, y, color='b', marker='o', mfc='r')

ax.set_xlabel('x轴坐标')
ax.set_ylabel('一些随机值')
```

## 坐标轴刻度

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei'] #解决中文乱码
plt.figure(figsize=(10, 5), edgecolor='r')

df=pd.read_excel('体温.xls')

x=df['日期']
y=df['体温']
plt.plot(x,y,color='m',linestyle='-',marker='o',mfc='w')
plt.xlabel('2020年2月')
plt.ylabel('基础体温')
# 图形里面的刻度是按照默认刻度显示的：2,4, 6...
```

```python
# 重新设置x轴刻度及标签
plt.figure(figsize=(10, 5), edgecolor='r')
dates=['1日','2日','3日','4日','5日',
       '6日','7日','8日','9日','10日',
       '11日','12日','13日','14日']
plt.xticks(range(1,15,1),dates)

plt.plot(x,y,color='m',linestyle='-',marker='o',mfc='w')
plt.xlabel('2020年2月')
plt.ylabel('基础体温')
```

## 坐标轴范围

## 设置网格线

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei'] #解决中文乱码
plt.figure(figsize=(10, 5), edgecolor='r')

dates=['1日','2日','3日','4日','5日',
       '6日','7日','8日','9日','10日',
       '11日','12日','13日','14日']
plt.xticks(range(1,15,1),dates)
plt.xlabel('2020年2月')
plt.ylabel('基础体温')
plt.grid(color='0.6', linestyle='--', linewidth=1, axis='x')

df=pd.read_excel('体温.xls')
x=df['日期']
y=df['体温']
plt.plot(x,y,color='m',linestyle='-',marker='o',mfc='w')
```

## 第二坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(7, 4)) # 设置图形大小

ax1 = fig.subplots()
np.random.seed(2000)
y = np.random.standard_normal((20, 2))
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro') # 设置颜色为红色，样式为圆形。
plt.legend(loc=8)

ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro') # 设置颜色为红色，样式为圆形。
plt.legend(loc=0) # 使用legend, loc设置位置
plt.ylabel('value 2nd') # 第二个y坐标

plt.grid(True) # 添加网格
plt.axis('tight') # 调整坐标宽度
```
