## 双坐标

- 使用`fig = Figure(figsize=(8, 4), dpi=100)`没有办法设置第二坐标
- 使用`plt.figure(figsize=(7, 4))`则可以

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import numpy as np

arr1 = np.linspace(0, 2, 100)
arr2 = np.random.randint(0, 100, 100)

#这行代码没有办法设置第二坐标
#fig = Figure(figsize=(8, 4), dpi=100)
#使用下面这行代码就可以。隐式接口`plt.plot()`会自动跟踪最近一次的axes，所以看起来ax2没有使用，实际上并非如此
fig = plt.figure(figsize=(7, 4)) # 设置图形大小
ax1 = fig.subplots()
plt.plot(arr1, 'b', lw=1.5, label='1ab')
plt.legend(loc=8)

ax2 = ax1.twinx()
plt.plot(arr2, 'g', lw=1.5, label='2ab')
plt.legend(loc=0)
```

```python
#单纯显示接口
fig, ax1 = plt.subplots(figsize=(7, 4))
ax1.plot(arr1, 'b', lw=1.5, label='1ab')
ax1.legend(loc=8)

ax2 = ax1.twinx()
ax2.plot(arr2, 'g', lw=1.5, label='2ab')
ax2.legend(loc=0)
```
