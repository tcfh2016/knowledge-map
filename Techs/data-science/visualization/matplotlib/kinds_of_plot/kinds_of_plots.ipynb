# 常用图表

## 折线图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_linechart = pd.read_excel('折线图数据.xls')
df_linechart
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

x1, y1, y2, y3 = df_linechart['姓名'], df_linechart['语文'], df_linechart['数学'], df_linechart['英语']

plt.plot(x1, y1, label='语文', color='r', marker='p')
plt.plot(x1, y2, label='数学', color='g', marker='.', mfc='r', ms=8, alpha=0.7)
plt.plot(x1, y3, label='英语', color='b', linestyle='-.', marker='*')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('books.xlsx')
x = df['年份']
height = df['销售额']
plt.bar(x, height, color='b', alpha=0.5, width=0.5, align='center')
plt.grid(axis='y', which='major')
plt.xlabel('年份')
plt.ylabel('线上销售额（元）')
plt.title('2013-2019年线上销售额分析图')
plt.legend(['销售额'], loc='upper left')

for a,b in zip(x, height):
    plt.text(a, b, format(b, ','), ha='center', va='bottom', fontsize=9, color='b', alpha=0.9)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('books.xlsx', sheet_name='Sheet2')
x = df['年份']
y1 = df['京东']
y2 = df['天猫']
y3 = df['自营']

plt.bar(x, y1, color='darkorange', alpha=0.5, width=0.25, align='center')
plt.bar(x+0.25, y2, color='deepskyblue', alpha=0.5, width=0.25, align='center')
plt.bar(x+0.50, y3, color='g', alpha=0.5, width=0.25, align='center')

plt.xlabel('年份')
plt.ylabel('线上销售额（元）')
plt.title('2013-2019年线上销售额分析图')
plt.legend(['京东', '天猫', '自营'], loc='upper left')

for a,b in zip(x, y1):
    plt.text(a, b, format(b, ','), ha='center', va='bottom', fontsize=9, color='b', alpha=0.9)
for a,b in zip(x, y2):
    plt.text(a+0.25, b, format(b, ','), ha='center', va='bottom', fontsize=9, color='b', alpha=0.9)
for a,b in zip(x, y3):
    plt.text(a+0.5, b, format(b, ','), ha='center', va='bottom', fontsize=9, color='b', alpha=0.9)
```

## 直方图

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2000)
y = np.random.standard_normal((1000, 2))

plt.hist(y, label=['1st', '2nd'], bins=25)

plt.grid(True) # 添加网格
plt.legend(loc=0)
plt.axis('tight') # 调整坐标宽度
plt.xlabel('value') # 设置x轴标签
plt.ylabel('frequency') # 设置x轴标签
plt.title('Histogram') # 设置图形标题
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

# 多子图

## `subplot()`

```python
import matplotlib.pyplot as plt
# 第一幅子图
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4, 5])
# 第二幅子图
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3, 4, 5], [2, 5, 8, 12, 15], 'ro')
# 第三幅子图，注意此时需要重新设置子图的布局，不再是之前的`(2,2)`，而变成了`(2,1)`
plt.subplot(2, 1, 2)
x = [1, 2, 3, 4, 5, 6]
h = [9, 5, 8, 12, 15, 20]
plt.bar(x, h)
```

## subplots()

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot([1, 2, 3, 4, 5])
axes[0, 1].plot([1, 2, 3, 4, 5], [2, 5, 8, 12, 15], 'ro')

x = [1, 2, 3, 4, 5, 6]
h = [9, 5, 8, 12, 15, 20]
axes[1, 0].bar(x, h)

axes[1, 1].pie([1, 2, 3, 4, 5], autopct='%1.1f%%')
```
