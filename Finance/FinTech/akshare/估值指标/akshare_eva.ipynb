## 获取A股多个估值指标

```python
import pandas as pd
import akshare as ak

indicator_dfs = []
for ind in ['总市值', '市盈率(TTM)', '市净率']:
    df = ak.stock_zh_valuation_baidu(symbol='603259', indicator=ind, period='全部')
    df = df.set_index('date').rename(columns={'value':ind})
    indicator_dfs.append(df)

df = pd.concat(indicator_dfs, axis=1)
df.plot(figsize=(12, 9), secondary_y=['市盈率(TTM)', '市净率'])
```

## 获取港股多个估值指标

```python
import pandas as pd
import akshare as ak
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.sans-serif'] = ['SimHei']

indicator_dfs = []
for ind in ['总市值', '市盈率(TTM)', '市净率']:
    df = ak.stock_hk_valuation_baidu(symbol='02618', indicator=ind, period='全部')
    df = df.set_index('date').rename(columns={'value':ind})
    indicator_dfs.append(df)

df = pd.concat(indicator_dfs, axis=1)
df.plot(figsize=(12, 9), secondary_y=['市盈率(TTM)', '市净率'])
```
